```python
new_df['overview_clean'] = new_df['overview'].str.lower().apply(lambda x: re.sub(r'[^a-zA-Z\s]','', x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(new_df['overview_clean'])
```

# Weighted Scoring

```python
new_df.describe()
```

```python
for i in new_df.select_dtypes(np.number).columns.to_list():
    print(i,'\n Skewness = ', new_df[i].skew(), '\n Kurtosis = ', new_df[i].kurt())
    print()
    print('**********************')
    print()
```

```python
new_df['popularity_log'] = np.log1p(new_df['popularity'])  # log(1 + x) to handle zeros
new_df['vote_count_log'] = np.log1p(new_df['vote_count'])
new_df['vote_average_log'] = np.log1p(new_df['vote_average'])
```

```python
for i in ['popularity_log', 'vote_count_log', 'vote_average_log']:
    print(i,'\n Skewness = ', new_df[i].skew(), '\n Kurtosis = ', new_df[i].kurt())
    print()
    print('**********************')
    print()
```

```python
new_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
m = MinMaxScaler()

new_df[['popularity_log', 'vote_count_log', 'vote_average_log']] = m.fit_transform(new_df[['popularity_log', 'vote_count_log', 'vote_average_log']])
```

```python
# new_df['vote_average_scaled'] = m.fit_transform(new_df[['vote_average']])
```

```python
new_df[["popularity_log", "vote_average_log", "vote_count_log"]].describe()
```

```python
def weighted_score(row, weight_popularity=0.5, weight_vote=0.3, weight_count=0.2):
    return(
    weight_popularity*row['popularity_log'] + 
        weight_vote*row['vote_average_log'] + 
        weight_count*row['vote_count_log']
    )
new_df['weighted_score'] = new_df.apply(weighted_score, axis=1)
```

```python
def get_recommendations(title, cosine_sim, new_df, top_n=5):
    idx = new_df[new_df['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity and apply weighted score
    sim_scores = sorted(
        sim_scores, key=lambda x: (x[1] * new_df.iloc[x[0]]['weighted_score']), reverse=True
    )
    
    top_indices = [i[0] for i in sim_scores[1:top_n + 1]]  # Exclude itself
    return new_df.iloc[top_indices][['title', 'overview','popularity_log', 'vote_average_log']]
```

```python
get_recommendations('The Crow', cosine_sim, new_df, top_n=5)
```
